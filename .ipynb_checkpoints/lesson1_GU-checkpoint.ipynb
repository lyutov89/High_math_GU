```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-np.pi, np.pi, 1000)
k1 = int(input('Введите коэффициент k  первой функции y=cos(kx)', ))
k2 = int(input('Введите коэффициент k второй функции y=cos(kx)', ))
plt.plot(x, np.cos(k1*x))
plt.plot(x, np.cos(k2*x))
plt.xlabel('x')
plt.ylabel('y')
plt.axis('tight')
plt.show()
```
