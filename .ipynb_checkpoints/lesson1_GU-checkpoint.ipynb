```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-np.pi, np.pi, 1000)
k = int(input('Введите коэффициент k функции y=cos(kx)', ))
plt.plot(x, np.cos(k*x))
plt.xlabel('x')
plt.ylabel('y')
plt.axis('tight')
plt.show()
```
