- Из урока по комбинаторике повторите расчеты, сгенерировав возможные варианты перестановок для других значений n и k

```python
n = 6 
s = ''
for i in range(0, 6):
    s += str(i)
print(f'Исходное сочетание: {s}')
```

```python
r=0
for p in itertools.permutations(s, 4):
    r+=1
    print(''.join(str(x) for x in p))
print(f'Кол-во перестановок элементов {r}')
print(f'По формуле выходит: {fact(n)/fact(n-4)}')
```

- Дополните код расчетом коэффициента корреляции x и y по формуле

```python
import matplotlib.pyplot as plt
import numpy as np
n=100
r=0.9
x=np.random.rand(n)
y = r*x + (1-r)*np.random.rand(n)
plt.plot(x,y,'o')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

```python
r = np.sum((x - np.mean(x))*(y - np.mean(y)))/np.sqrt(np.sum((x - np.mean(x))**2) * np.sum((y - np.mean(y))**2))
```
