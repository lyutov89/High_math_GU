Поверхности второго порядка. 

```python
from mpl_toolkits.mplot3d import Axes3D  

import matplotlib.pyplot as plt
import numpy as np
%config InlineBackend.figure_format = 'svg'
%matplotlib inline

fig = plt.figure()
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
Z = X**2 + Y**2
ax=Axes3D(fig)
ax.plot_surface(X, Y, Z, rstride=1, cstride=1)

plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D  

import matplotlib.pyplot as plt
import numpy as np
%config InlineBackend.figure_format = 'svg'
%matplotlib inline

fig = plt.figure()
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
Z = X**2 - Y**2
ax=Axes3D(fig)
ax.plot_surface(X, Y, Z, rstride=5, cstride=5)

plt.show()

```

```python
from mpl_toolkits.mplot3d import Axes3D  
from matplotlib import cm
import matplotlib.pyplot as plt
import numpy as np
%config InlineBackend.figure_format = 'svg'
%matplotlib inline

fig = plt.figure()
X = np.arange(-20, 20, 2)
Y = np.arange(-20, 20, 2)
X1 = np.arange(-20, 20, 2)
Y1 = np.arange(-20, 20, 2)
X, Y = np.meshgrid(X, Y)
X1, Y1 = np.meshgrid(X1, Y1)
Z = X - Y
Z1 = X1 - Y1 + 15
ax=Axes3D(fig)
ax.plot_surface(X, Y, Z, rstride=1, cstride=4, cmap = cm.plasma)
ax.plot_surface(X1, Y1, Z1, rstride=1, cstride=4)

plt.show()
```

y(x) = k∙cos(x – a) + b
для некоторых (2-3 различных) значений параметров k, a, b


```python
import matplotlib.pyplot as plt
import numpy as np
%config InlineBackend.figure_format = 'svg'
%matplotlib inline

x=np.linspace(-10, 10, 2000)

k = int(input('Введите коэффициент k: '))
a = int(input('Введите коэффициент a: '))
b = int(input('Введите коэффициент b: '))
y=k*np.cos(x-a)+b
plt.plot(x,y)
plt.gca().set_aspect('equal')
plt.grid(True)
plt.show()
```

Так как polar plots используют полярную систему координат, кратко вспомним что это такое и как она связана с декартовой (прямоугольной) системой координат.

В декартовой системе координат, которая используется по умолчанию, положение точки определяется двумя значениями: аргументом x и значением функции y.

В полярной системе координат положение точки также определяется двумя значениями: радиусом r и углом phi. Угол обычно отсчитывают против часовой стрелке от "восточного" (3:00) направления.

Переход из полярной системы координат в прямоугольную выглядит так:

x = r*cos(phi)

y = r*sin(phi)

```python
phi = np.linspace(0, 2*np.pi, 1000)
r=5 
x = r*np.cos(phi)
y = r*np.sin(phi)
plt.plot(x,y)
plt.grid(True)
```

```python
from scipy.optimize import fsolve
x=np.linspace(-5,5,201)
plt.plot(x, x**2 - 1)
plt.plot(x, (np.exp(x) + x - 1)/x)
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
```

```python
def equations(p):
    x,y = p
    return(y - x**2 + 1, np.exp(x) + x*(1 - y) - 1)

x1, y1 = fsolve(equations, (-2,-1))
x2, y2 = fsolve(equations, (2,3))
x3, y3 = fsolve(equations, (3.5,4.5))

print(x1, y1)
print(x2, y2)
print(x3, y3)
```
